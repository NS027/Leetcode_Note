### âœ… Cè¯­è¨€å®ç°ï¼ˆç”¨æ•°ç»„æ¨¡æ‹Ÿæ ˆï¼‰

```c
#define MAX_SIZE 100

void preorderTraversal(struct TreeNode* root, int* returnArray, int* returnSize) {
    if (root == NULL) return;

    struct TreeNode* stack[MAX_SIZE];
    int top = -1;
    stack[++top] = root;

    while (top >= 0) {
        struct TreeNode* node = stack[top--];
        returnArray[(*returnSize)++] = node->val;

        if (node->right) stack[++top] = node->right;
        if (node->left)  stack[++top] = node->left;
    }
}
```

```python
def inorderTraversal(root):
    result = []
    stack = []
    current = root

    while current or stack:
        while current:
            stack.append(current)
            current = current.left

        current = stack.pop()
        result.append(current.val)
        current = current.right

    return result
```

### âœ… Java å®ç°

```java
public List<Integer> inorderTraversal(TreeNode root) {
    List<Integer> result = new ArrayList<>();
    Stack<TreeNode> stack = new Stack<>();
    TreeNode curr = root;

    while (curr != null || !stack.isEmpty()) {
        while (curr != null) {
            stack.push(curr);
            curr = curr.left;
        }

        curr = stack.pop();
        result.add(curr.val);
        curr = curr.right;
    }

    return result;
}
```

```python
from collections import deque

def levelOrder(root):
    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        size = len(queue)
        level = []

        for _ in range(size):
            node = queue.popleft()
            level.append(node.val)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        result.extend(level)  # å¦‚æœåªæƒ³è¦å¹³é“ºè¾“å‡º

    return result
```

### âœ… Java å®ç°

```java
import java.util.*;

public class Solution {
    public List<Integer> levelOrder(TreeNode root) {
        List<Integer> result = new ArrayList<>();
        if (root == null) return result;

        Queue<TreeNode> queue = new LinkedList<>();
        queue.offer(root);

        while (!queue.isEmpty()) {
            int size = queue.size();

            for (int i = 0; i < size; i++) {
                TreeNode node = queue.poll();
                result.add(node.val);

                if (node.left != null) queue.offer(node.left);
                if (node.right != null) queue.offer(node.right);
            }
        }

        return result;
    }
}
```
